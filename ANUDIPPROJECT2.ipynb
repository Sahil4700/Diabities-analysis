```python
#1.Bar Plot: Class Distribution (Diabetic vs Non-Diabetic)
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
df = pd.read_csv('diabetes_data_upload.csv')
sns.set(style="whitegrid")  # Optional: improves visual style
plt.figure(figsize=(6, 4))  # Set figure size
sns.countplot(data=df, x='class', palette='Set2')
plt.title('Diabetes Class Distribution')
plt.xlabel('Diabetes Status')
plt.ylabel('Number of Patients')
plt.tight_layout()
plt.show()
```

```python
#2 calculate the gender distribution, and plot the pie chart:
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('diabetes_data_upload.csv')
# Gender distribution counts
gender_counts = df['Gender'].value_counts()
# Plot pie chart
plt.figure(figsize=(6,6))
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', colors=['lightblue', 'lightgreen'])
plt.title('Gender Distribution')
plt.show()
```

```python
#3  Histogram Showing the Age Distribution of Patients
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
df = pd.read_csv('diabetes_data_upload.csv')
sns.set(style="whitegrid")
plt.figure(figsize=(8, 5))
# Histogram of Age distribution
sns.histplot(data=df, x='Age', bins=20, kde=True, color='lightgreen')
plt.title('Distribution of Patient Ages')
plt.xlabel('Age')
plt.ylabel('Number of Patients')
plt.tight_layout()
plt.show()
```
